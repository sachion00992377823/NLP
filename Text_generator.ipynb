```python
import requests
from PIL import Image
from io import BytesIO
from IPython.display import display
import os

API_URL = "https://api-inference.huggingface.co/models/stabilityai/stable-diffusion-2"

# Retrieve the Hugging Face token from Colab secrets or environment variables
hf_token = os.getenv("HF_TOKEN")

if hf_token is None:
    print("Warning: HF_TOKEN not found. Please set your Hugging Face token in Colab secrets or environment variables.")
    # For demonstration purposes, you might want to exit or handle this more gracefully.
    # For now, we'll proceed, but API calls will likely fail without a valid token.
    headers = {"Authorization": "Bearer YOUR_HF_TOKEN_IS_MISSING"}
else:
    headers = {"Authorization": f"Bearer {hf_token}"}

def generate_image(prompt):
    response = requests.post(API_URL, headers=headers, json={"inputs": prompt})
    # Check for successful response before trying to open content
    response.raise_for_status() # Raise an exception for HTTP errors (4xx or 5xx)
    image = Image.open(BytesIO(response.content))
    return image

try:
    image = generate_image("a circuit board glowing with blue light")
    image.save("generated.png")
    display(image)
except requests.exceptions.RequestException as e:
    print(f"An error occurred during the API call: {e}")
    print("Please ensure your HF_TOKEN is correctly set and you have network access.")
```

```python
# @title AI prompt cell

import ipywidgets as widgets
from IPython.display import display, HTML, Markdown,clear_output
from google.colab import ai

dropdown = widgets.Dropdown(
    options=[],
    layout={'width': 'auto'}
)

def update_model_list(new_options):
    dropdown.options = new_options
update_model_list(ai.list_models())

text_input = widgets.Textarea(
    placeholder='Ask me anything....',
    layout={'width': 'auto', 'height': '100px'},
)

button = widgets.Button(
    description='Submit Text',
    disabled=False,
    tooltip='Click to submit the text',
    icon='check'
)

output_area = widgets.Output(
     layout={'width': 'auto', 'max_height': '300px','overflow_y': 'scroll'}
)

def on_button_clicked(b):
    with output_area:
        output_area.clear_output(wait=False)
        accumulated_content = ""
        for new_chunk in ai.generate_text(prompt=text_input.value, model_name=dropdown.value, stream=True):
            if new_chunk is None:
                continue
            accumulated_content += new_chunk
            clear_output(wait=True)
            display(Markdown(accumulated_content))

button.on_click(on_button_clicked)
vbox = widgets.GridBox([dropdown, text_input, button, output_area])

display(HTML("""
<style>
.widget-dropdown select {
    font-size: 18px;
    font-family: "Arial", sans-serif;
}
.widget-textarea textarea {
    font-size: 18px;
    font-family: "Arial", sans-serif;
}
</style>
"""))
display(vbox)
```
